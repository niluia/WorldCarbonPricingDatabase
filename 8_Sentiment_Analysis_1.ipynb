```python
from sklearn.feature_extraction.text import CountVectorizer # It is making a Bag of Word
```

```python
cv = CountVectorizer(max_features = 1500) # We are creating the WordBag from the 1500 most common words.
```

A fit_transform metódus először megtanulja a szövegben található szavakat és azok gyakoriságát (ez a "fit" rész), majd átalakítja a szövegeket numerikus vektorokká. Minden vektor egy dokumentumot reprezentál, és az egyes elemek a megfelelő szavak gyakoriságát jelzik (ez a "transform" rész).

```python
X = cv.fit_transform(corpus).toarray()
```

```python
y = dataset.iloc[:, -1].values # This code loads the target variable into the y variable. The dataset.iloc[:, -1] selects the last column of the dataset.
```

```python
from sklearn.naive_bayes import GaussianNB
# Training the Naive Bayes Model on the Training Set
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Alkalmazzuk az előfeldolgozási lépéseket
def preprocess_review(text):
    # Remove non-letter characters
    text = re.sub('[^a-zA-Z]', ' ', text)
    # Convert to lowercase
    text = text.lower()
    # Split into words
    text = text.split()
    # Apply stemming and remove stopwords
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')  # Keep the word 'not'
    text = [ps.stem(word) for word in text if not word in set(all_stopwords)]
    # Join the words back into one string
    return ' '.join(text)
```

```python
# Classifying a New Review Based on Our Model
new_review = "Overall, dining at this restaurant was a delight from start to finish. The delicious food, impeccable service, and charming ambiance made for a memorable experience."
```

```python
processed_review = preprocess_review(new_review)
```

```python
print(processed_review)
```

```python
# Using CountVectorizer to Transform the New Review
new_review_transformed = cv.transform([processed_review]).toarray()
```

```python
# Using the Trained GaussianNB Classifier on the New Review
classification = classifier.predict(new_review_transformed)
```

```python
# Displaying the Result
if classification == 1:
    print(f"The review is positive: {new_review}")
else:
    print(f"The review is negative: {new_review}")
```
